```python
def find_largest(num1, num2, num3):
    if (num1 >= num2) and (num1 >= num3):
        largest = num1
    elif (num2 >= num1) and (num2 >= num3):
        largest = num2
    else:
        largest = num3
    print(f"The largest number among {num1}, {num2}, and {num3} is: {largest}")
```

```python
text = "hello"
vowels = ['a', 'e', 'i', 'o', 'u']

def count(text):
    v = 0
    for i in text:
        if i in vowels:
            v += 1
    return v

print(count(text))
```

```python
def minimumSwap(s1, s2):
    x = y = 0

    for a, b in zip(s1, s2):
        if a == '0' and b == '1':
            x += 1
        elif a == '1' and b == '0':
            y += 1

    if (x + y) % 2 != 0:
        return -1

    return x//2 + y//2 + 2*(x % 2)
```
